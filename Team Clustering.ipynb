```python
def find_average_scores(a, b):
    avg_arr=[]
    for i in range(0,5):
        avg = (a[i]+b[i])/2
        avg_arr.append(avg)
    return avg_arr
```

```python
def team_combination(final_score, next_batch):
    combo_arr=[]
    for i in range(0, len(final_score)):
        combo_arr.append([])
    for i in range (0,len(final_score)):
        for j in range(0,len(next_batch)):
            x = find_average_scores(final_score[i], next_batch[j])
            combo_arr[i].append(x)
    return combo_arr
```

```python
def best_team(matrix):
    combinations = []
    indexes=[]
    for combination in itertools.combinations(range(n * n), n):
        row_indices = [i // n for i in combination] 
        col_indices = [i % n for i in combination]

        if len(set(row_indices)) == n and len(set(col_indices)) == n:
            is_valid = all(matrix[i // n][i % n] != matrix[j // n][j % n] for i, j in itertools.combinations(combination, 2))
            if is_valid:
                combinations.append([matrix[i // n][i % n] for i in combination])
                indices = [[i // n, i % n] for i in combination]
                indexes.append(indices)
    std_arr = []
    for i in range(0, len(combinations)):
        sum_arr = []
        for j in range(0, len(combinations[i])):
            total = 0
            for k in range(0, len(combinations[i][j])):
                total += combinations[i][j][k]
            
            sum_arr.append(total)
        std_arr.append(np.std(sum_arr))
    min_std = min(std_arr)
    min_std_index = std_arr.index(min_std)
    best_combo=indexes[min_std_index]
    return best_combo
```

```python
def run_code_with_n(n):
    start_time = time.time()
    
    final_team = []
    for i in range(n):
        final_team.append([])

    final_score = []
    for i in range(n):
        final_score.append([responses_df["Openness "][i], responses_df["Conscientiousness "][i], responses_df["Extraversion "][i], responses_df["Agreeableness "][i], responses_df["Neuroticism "][i]])

    # Code for leader
    for i in range(0, n):
        final_team[i].append(responses_df["Name"][i])

    for i in range(n, len(responses_df) - len(responses_df) % n, n):
        next_batch = []
        names_next_batch = []
        for j in range(i, i + n):
            next_batch.append([responses_df["Openness "][j], responses_df["Conscientiousness "][j], responses_df["Extraversion "][j], responses_df["Agreeableness "][j], responses_df["Neuroticism "][j]])
            names_next_batch.append(responses_df["Name"][j])
            team_combos = team_combination(final_score, next_batch)
        best_combo = best_team(team_combos)
        # Adding the best match
        for i in range(0, len(final_score)):
            final_index = best_combo[i][0]
            next_batch_index = best_combo[i][1]
            new_score = find_average_scores(final_score[final_index], next_batch[next_batch_index])
            final_score[final_index] = new_score
            final_team[final_index].append(names_next_batch[next_batch_index])

    end_time = time.time()
    execution_time = end_time - start_time
    return execution_time
```

```python
n_values = list(range(5, 101, 5))
execution_times = []
```

```python
for n in n_values:
    execution_time = run_code_with_n(n)
    execution_times.append(execution_time)
```

```python
plt.plot(n_values, execution_times, marker='o')
plt.xlabel('n')
plt.ylabel('Execution Time (seconds)')
plt.title('n vs. Execution Time')
plt.grid()
plt.show()
```
